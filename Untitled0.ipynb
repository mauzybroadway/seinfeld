```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('./mag_data.csv', delimiter=',', skip_header=10,
                     skip_footer=10, names=['x', 'y', 'z'])
```

```python
def plot_bounds(a, b):
    fig = plt.figure()

    ax1 = fig.add_subplot(111)

    ax1.set_title("Mains power stability")    
    ax1.set_xlabel('x')
    ax1.set_ylabel('z')

    dat = data[a:b]

    ax1.plot(dat['x'], dat['z'], color='r', label='the data')

    leg = ax1.legend()

    plt.show()
```

```python
plot_bounds(0,15)
```

```python
plot_bounds(30,55)
```

```python
plot_bounds(70,85)
```

```python
plot_bounds(100,115)
```

```python
plot_bounds(130,150)
```

```python
angle = np.sqrt(data['x']**2 + data['z']**2)
```

```python
def plot_direction(a, b):
    fig = plt.figure()

    ax1 = fig.add_subplot(111)

    ax1.set_title("Mains power stability")    
    ax1.set_xlabel('x')
    ax1.set_ylabel('z')

    ang = angle[a:b]

    ax1.plot(ang, color='r', label='the data')

    leg = ax1.legend()

    plt.show()
```

```python
plot_direction(0,len(angle))
```
